```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer los datos de los archivos CSV
data_v1 = pd.read_csv('drive/MyDrive/Glottis/colab_glottis/graficas/model_GlottisV1.csv')
data_v2e = pd.read_csv('drive/MyDrive/Glottis/colab_glottis/graficas/model_GlottisV2e.csv')

# Filtrar los DataFrames para incluir solo los datos hasta la epoch 10
data_v1 = data_v1[data_v1['epoch'] <= 10]
data_v2e = data_v2e[data_v2e['epoch'] <= 10]

# Graficar las mtricas de los dos modelos
plt.figure(figsize=(10, 5))
plt.plot(data_v1['epoch'], data_v1['val_ap_pred_MAPE_V1'], label='Modelo GlottisV1 (hasta epoch 10)', color='magenta')
plt.plot(data_v2e['epoch'], data_v2e['val_ap_pred_metric_mape'], label='Modelo GlottisV2e (hasta epoch 10)', color='green')

# Nombrar los ejes y agregar ttulo y leyenda
plt.xlabel('Epochs')
plt.ylabel('AP Pred Metric MAPE')
plt.title('Comparacin de la mtrica AP Pred Metric MAPE hasta Epoch 10')
plt.legend()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer los datos de los archivos CSV
data_v1 = pd.read_csv('drive/MyDrive/Glottis/colab_glottis/graficas/model_GlottisV1.csv')
data_v2e = pd.read_csv('drive/MyDrive/Glottis/colab_glottis/graficas/model_GlottisV2e.csv')

# Filtrar los DataFrames para incluir solo los datos hasta la epoch 10
data_v1 = data_v1[data_v1['epoch'] <= 10]
data_v2e = data_v2e[data_v2e['epoch'] <= 10]

# Graficar las mtricas de los dos modelos
plt.figure(figsize=(10, 5))

# Para el modelo con 20 epochs, usamos color rojo
plt.plot(data_v1['epoch'], data_v1['val_seg_iou_score'], label='Modelo GlottisV1 (10 epochs)', color='magenta')

# Para el modelo con 30 epochs, usamos color azul
plt.plot(data_v2e['epoch'], data_v2e['val_seg_iou_score'], label='Modelo GlottisV2e (10 epochs)', color='green')

# Nombrar los ejes
plt.xlabel('Epochs')
plt.ylabel('IoU socre')

# Ttulo y leyenda
plt.title('Comparacin de la mtrica IoU por Epochs')
plt.legend()

# Mostrar el grfico
plt.show()
```
